Wyodrębnianie poszczególnych opinii

```python
page_dom = BeautifulSoup(response.text, "html.parser")
reviews = page_dom.select("div.js_product-review:not(.user-post--highlight)") # pobranie wszystkich opinii
```

Wyciąganie opinii

```python
all_opinions = []
for review in reviews:
    single_opinion = {
        "opinion_id": review["data-entry-id"],
        "author": review.select_one("span.user-post__author-name").text.strip() if review.select_one("span.user-post__author-name") else "Unknown Author",
        "recommendation": review.select_one("span.user-post__author-recomendation em").text.strip() if review.select_one("span.user-post__author-recomendation em") else "No Recommendation",
        "rating": review.select_one("span.user-post__score-count").text.strip() if review.select_one("span.user-post__score-count") else "No Rating",
        "content": review.select_one("div.user-post__text").text.strip() if review.select_one("div.user-post__text") else "No Content",
        "pros": [pros.text.strip() for pros in review.select("div.review-feature__section .review-feature__item--positive")],
        "cons": [cons.text.strip() for cons in review.select("div.review-feature__section .review-feature__item--negative")],
        "helpful": review.select_one("button.vote-yes")["data-total-vote"] if review.select_one("button.vote-yes") else "0",
        "nothelpful": review.select_one("button.vote-no")["data-total-vote"] if review.select_one("button.vote-no") else "0",
        "published_date": review.select_one("span.user-post__published time").text.strip() if review.select_one("span.user-post__published time") else "Unknown Date",
        "purchase_date": review.select_one("span.user-post__purchased").text.strip() if review.select_one("span.user-post__purchased") else "Unknown Purchase Date",
    }
    all_opinions.append(single_opinion)
    
    print(single_opinion)
```

Zapisanie do pliku JSON

```python
with open("opinions.json", "w", encoding="utf-8") as file:
    import json
    json.dump(all_opinions, file, ensure_ascii=False, indent=4)
```
