#### WEB Scraper - 
https://books.toscrape.com/

```python
import json
import requests
from bs4 import BeautifulSoup
import time
from urllib.parse import urljoin
```

```python
# Zaczynamy od pierwszej strony katalogu
current_url = "https://books.toscrape.com/catalogue/page-1.html"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36"
}

all_books_data = []
page_number = 1

print("Rozpoczynam pobieranie całej biblioteki...\n")

# Pętla wykonuje się tak długo, jak długo zmienna current_url nie jest pusta
while current_url:
    print(f"Skupiam się na stronie {page_number}: {current_url}")
    response = requests.get(current_url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        books = soup.select("article.product_pod")
        
        # Pobieranie danych o książkach z bieżącej strony
        for book in books:
            title = book.select_one("h3 a").get("title")
            price = book.select_one("p.price_color").text.strip()
            availability = book.select_one("p.instock.availability").text.strip()
            
            # urljoin automatycznie łączy główny człon domeny z linkiem do książki
            link_fragment = book.select_one("h3 a").get("href")
            full_link = urljoin(current_url, link_fragment)
            
            all_books_data.append({
                "title": title,
                "price": price,
                "availability": availability,
                "url": full_link
            })
            
        # Szukanie przycisku "Next" prowadzącego do następnej strony
        next_button = soup.select_one("li.next a")
        
        if next_button:
            # Jeśli przycisk istnieje, budujemy nowy adres URL i kręcimy pętlę dalej
            next_url_fragment = next_button.get("href")
            current_url = urljoin(current_url, next_url_fragment)
            page_number += 1
            
            # DOBRA PRAKTYKA: Dodajemy sekundę pauzy, żeby nie bombardować serwera zapytaniami
            time.sleep(1) 
        else:
            # Jeśli przycisku "Next" nie ma, to znaczy, że jesteśmy na ostatniej stronie
            print("\nDotarłem do ostatniej strony!")
            current_url = None # To zakończy pętlę while
            
    else:
        print(f"Wystąpił błąd na stronie {page_number}. Kod statusu: {response.status_code}")
        break

print(f"\nUkończono! Pobrano dane o łącznie {len(all_books_data)} książkach.")

# Zapis do pliku
output_file = "wszystkie_ksiazki_toscrape.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(all_books_data, f, indent=4, ensure_ascii=False)

print(f"Dane zostały bezpiecznie zapisane w pliku: {output_file}")
```
